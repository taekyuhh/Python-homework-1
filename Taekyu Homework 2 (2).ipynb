# 문제1) 우리는 지난 시간에 setosa의 sepal length에 대해서 하나씩 그래프를 그렸습니다. 다른 iris type의 다른 feature에 대해 똑같이 그래프를 그려볼까요?

## Bringing Data


```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
feature = iris['data']
```

```python
labels = iris['target']
```

```python
feature_names = iris['feature_names']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(feature, columns=feature_names)
```

```python
df['target'] = labels
```

```python
df0 = df[df['target'] ==0]
df1 = df[df['target'] ==1]
df2 = df[df['target'] ==2]
```

```python
import matplotlib.pyplot as plt
```

## Versicolor petal length

### Line

```python
plt.figure(figsize=(6,4))

plt.plot(df1['petal length (cm)'])

plt.title('Petal length line graph of iris versicolor', fontsize=14)

plt.show()
```

### Bar

```python
import numpy as np
```

```python
np.arange(len(df1))
```

```python
plt.figure(figsize=(10,4))

#There needs to be an x value
x=np.arange(len(df1))
plt.bar(x, df1['petal length (cm)'])
plt.title('Petal length of iris versicolor', fontsize=10)
plt.show()
```

### Box Plots

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()
ax.boxplot(df1['petal length (cm)'])
plt.title('Petal length of iris versicolor', fontsize=20)
plt.show()
```

### Pie

```python
plt.figure(figsize = (6,4))
ratio = [len(df0), len(df1), len(df2)]
labels = ['setosa', 'versicolor', 'verginica']

plt.pie(ratio,labels=labels, autopct='%.1f%%')

plt.legend(bbox_to_anchor=(0,0))

plt.show()
```

# 문제2) 우리는 지난 시간에 모든 iris의 sepal length에 대해서 그래프를 겹쳐 그렸습니다. 다른 feature에 대해 똑같이 그래프를 그려볼까요?

```python
plt.figure(figsize=(8,4))

plt.hist(df0['sepal width (cm)'])
plt.hist(df1['sepal width (cm)'])
plt.hist(df2['sepal width (cm)'])

plt.title('Sepal Width Histograms of 3 types of iris', fontsize=14)
plt.show()
```

# 문제3) 새로운 와인 데이터를 다뤄봅시다. 
## feature 중 한 가지를 골라서 세 가지의 class를 동시에 그리는 pie chart, box plot, histogram을 그려봅시다. 
## 우리가 그렸던 그래프와 같이 legend도 포함해주세요.

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine
```

```python
feature_w = wine['data']
```

```python
labels_w = wine['target']
```

```python
feature_names_w = wine['feature_names']
```

```python
feature_names_w
```

```python
dfw = pd.DataFrame(feature_w, columns=feature_names_w)
```

```python
dfw
```

```python
dfw0 = dfw[dfw['magnesium']>=120]
dfw1 = dfw[(dfw['magnesium']<=120) & (dfw['magnesium']>100)]
dfw2 = dfw[dfw['magnesium']<=100]
```

## 3 graphs

```python
plt.figure(figsize=(6,4))
fig, ax = plt.subplots()

ax.boxplot([dfw0['magnesium'], dfw1['magnesium'], dfw2['magnesium']])
plt.legend
plt.show()
```

```python
plt.figure(figsize = (6,4))

ratio = [len(dfw0), len(dfw1), len(dfw2)]
labels = ['120+', '120-100', 'below 100']
plt.pie(ratio,labels=labels, autopct='%.1f%%')
plt.legend(bbox_to_anchor=(0,0))
plt.show()
```

```python
plt.figure(figsize = (6,4))

plt.hist(dfw0['magnesium'], label = '120+' )
plt.hist(dfw1['magnesium'], label = '100-120')
plt.hist(dfw2['magnesium'], label = 'below 100')

plt.legend()

plt.show()
```
